```python
#Exo 4
def missingValueByColumn(dataFrame):
    null_columns=dataFrame.columns[dataFrame.isnull().any()]
    print(dataFrame[null_columns].isnull().sum())

missingValueByColumn(df)
```

```python
#Exo 5
def replaceMissingValue(dataFrame,columnName):
    values=dataFrame[columnName].mean
    dataFrame[columnName].apply(pd.to_numeric, errors='coerce')
    dataFrame.fillna(value=values, inplace=True)
    return dataFrame

replaceMissingValue(df,'prod_cost')
```

```python
#Exo 6
def uniqueValue(dataFrame,columnName):
    return dataFrame[columnName].unique()

uniqueValue(df,'warranty')
```

```python
#Exo 7,8
def numericValue(dataFrame,columnName):
    return dataFrame[columnName].str[0].unique()

numericValue(df,'warranty')
```

```python
#Exo 9
def subCategory(dataFrame, column, values):
    for i in range (len(dataFrame)):
            if type(dataFrame.loc[i, column]) != str :
                if dataFrame.loc[i, column] < 300 :
                    dataFrame.loc[i, column] = values[0]
                elif dataFrame.loc[i, column] > 300 and dataFrame.loc[i, column] < 500 :
                    dataFrame.loc[i, column] = values[1]
                else :
                    dataFrame.loc[i, column] = values[2]
      
    
values = ['0-300', '301-500', '501++']
subCategory(df, 'price', values)
df
```
